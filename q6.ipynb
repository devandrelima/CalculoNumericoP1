```python
import math
import numpy as np
import time
# Função f(x)
def f(x):
    return x**3 - x - 1

# Função de iteração g(x)
def g(x):
    return np.cbrt(x + 1)

# Método do Ponto Fixo
def ponto_fixo(a, tol):
    iter_count = 0
    c = a  # Ponto inicial
    gc = g(c)

    while abs(f(gc)) > tol:
        c = gc
        gc = g(c)
        print(f"Iteração {iter_count}: x = {c}, f = {gc}")
        iter_count += 1
    return gc

############### Método da bissecção #####################
def calcK(a, b, epsilon):
    return math.ceil(math.log2((abs(b-a))/epsilon))

def bisseccao(a, b, epsilon, k):
    iter_count = 0
    # Para intervalos com f(a) e f(b) com o mesmo sinal
    if f(a) * f(b) >= 0:
        while (b - a) > epsilon:
            x = (a + b) / 2 # Calculo do ponto médio
            FX = f(x) # Calculo de f(x)
            FA = f(a) # Calculo de f(a)
            FB = f(b) # Calculo de f(b)
            # Verificando se os valores no intervalo são menores que epsilon e consequentemente a raiz
            print(f"Iteração {iter_count}: x = {x}, f(x) = {f(x)}")
            if abs(FA) < epsilon:
                return a
            elif abs(FB) < epsilon:
                return b
            # Verificando se o valor de f(x) é menor que epsilon e consequentemente a raiz
            elif abs(FX) < epsilon:
                return x
            # Escolha do intervalo
            elif FB > FA:
                b = x # Se f(b) for maior que f(a) então o intervalo é [a, x]
            else:
                a = x # Se f(a) for maior que f(b) então o intervalo é [x, b]
            iter_count += 1
            if iter_count > k: # Se o número de iterações for maior que k, então não foi possível encontrar a raiz
                return None
    else:
        # Para intervalos com f(a) e f(b) opostos
        while abs(b - a) > epsilon:
            x = (a + b) / 2
            if f(x) == 0:
                return x
            elif f(a) * f(x) < 0:
                b = x
            else:
                a = x
            iter_count += 1
            print(f"Iteração {iter_count}: x = {x}, f(x) = {f(x)}")
        return x


a = float(input("Digite o valor de a: "))
b = float(input("Digite o valor de b: "))
epsilon = float(input("Digite o valor da tolerância: "))
tolerancia = epsilon 
k = calcK(a, b, epsilon)

print("#### Método da bissecção ####")
inicio = time.time()
raizB = bisseccao(a, b, epsilon, k)
fim = time.time()
if raizB is not None:
    print(f"\nRaiz aproximada: {raizB:.5f} em {fim - inicio:.5f} segundos.")
else:
    print("\nNão foi possível encontrar a raiz.")   

print("\n#### Método do Ponto Fixo ####")
inicio = time.time()
raiz_aproximada = ponto_fixo(a, tolerancia)
fim = time.time()
if raiz_aproximada is not None:
    print(f"\nRaiz aproximada: {raiz_aproximada:.5f} em {fim - inicio:.5f} segundos.")
else: 
    print("\nNão foi possível encontrar a raiz.")
```
