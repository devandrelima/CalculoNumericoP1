```python
import time

def f(x):
    return x**3 - x - 1

def derivada(x, epsilon):
    return (f(x + epsilon) - f(x)) / epsilon

def metodo_newton(x0, epsilon, max_iter=100):
    time.sleep(1)
    x = x0
    iter_count = 0
    print("------------------- Método de Newton -------------------")
    while abs(f(x)) > epsilon and iter_count < max_iter:
        df = derivada(x, epsilon)
        if df == 0:
            raise ValueError("A derivada é zero. O método de Newton pode falhar.")

        x = x - f(x) / df
        iter_count += 1
        print(f"Iteração {iter_count}: x = {x}, f(x) = {f(x)}")

        if iter_count == max_iter:
            return None
    return x

def metodo_secante(x0, x1, epsilon, max_iter=100):
    time.sleep(1)

    iter_count = 0
    print("------------------- Método da Secante -------------------")
    while abs(f(x1)) > epsilon and iter_count < max_iter:
        if f(x1) == f(x0):
            raise ValueError("Divisão por zero na iteração. Método falhou.")

        # Fórmula do método da secante
        x_new = x1 - f(x1) * (x1 - x0) / (f(x1) - f(x0))

        # Atualizar os pontos
        x0, x1 = x1, x_new
        iter_count += 1
        print(f"Iteração {iter_count}: x = {x1}, f(x) = {f(x1)}")

    if iter_count == max_iter:
        print("Número máximo de iterações atingido.")
        return None

    return x1

# Entrada do usuário
x0 = float(input("Digite o ponto inicial (Newton/Secante): "))
x1 = float(input("Digite o segundo ponto inicial (x1) (Secante): "))
epsilon = float(input("Digite o valor da precisão (Newton/Secante): "))

# Encontrando a raiz
inicioN = time.time()
raizN = metodo_newton(x0, epsilon)
fimN = time.time() - 1

inicioS = time.time()
raizS = metodo_secante(x0, x1, epsilon)
fimS = time.time() - 1

if raizN and raizS is not None: 
    print(f"Raiz Newton: {raizN} em {(fimN - inicioN):.5f} segundos.\nRaiz Secante: {raizS} em {(fimS - inicioS):.5f} segundos.")
else:
    print(f"O código de Newton executou em {(fimN - inicioN):.5f} segundos, mas não foi possível encontrar a raiz com os parâmetros fornecidos.")
```
