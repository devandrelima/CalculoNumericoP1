```python
## Q4
#####
import math
import time

# Função
def f(x):
    return math.exp(x**2) + 3*x

def calcFakePosition(a, b):
  return (a * f(b) - b * f(a)) / (f(b) - f(a))
  
iter_count = 0

# Método da bissecção
def falsa_posicao(a, b, epsilon, k):
    global iter_count
    # Para intervalos com f(a) e f(b) com o mesmo sinal
    try:
        while abs(b - a) > epsilon and iter_count < k:
            x = calcFakePosition(a, b)
            FX = f(x) # Calculo de f(x)
            FA = f(a) # Calculo de f(a)
            FB = f(b) # Calculo de f(b)
            print(x, FX)
            
            # Verificando se os valores no intervalo são menores que epsilon e consequentemente a raiz
            if abs(FA) < epsilon:
                return a
            elif abs(FB) < epsilon:
                return b
            # Verificando se o valor de f(x) é menor que epsilon e consequentemente a raiz
            elif abs(FX) < epsilon:
                return x
            # Escolha do intervalo
            elif FX * FA < 0:
                b = x # Se f(b) for maior que f(a) então o intervalo é [a, x]
            else:
                a = x # Se f(a) for maior que f(b) então o intervalo é [x, b]

            iter_count += 1
    except:
        print("Divisão por zero")
        return None

# Calculo de interações
def calcK(a, b, epsilon):
    return math.ceil(math.log2((abs(b-a))/epsilon))

iter_count_bisse = 0

# Método da bissecção
def bisseccao(a, b, epsilon, k):
    time.sleep(1)
    global iter_count_bisse
    # Para intervalos com f(a) e f(b) com o mesmo sinal
    if f(a) * f(b) >= 0:
        while abs(b - a) > epsilon:
            x = (a + b) / 2 # Calculo do ponto médio
            FX = f(x) # Calculo de f(x)
            FA = f(a) # Calculo de f(a)
            FB = f(b) # Calculo de f(b)
            # Verificando se os valores no intervalo são menores que epsilon e consequentemente a raiz
            print(f"Iteração {iter_count_bisse}: x = {x}, f(x) = {f(x)}")
            if abs(FA) < epsilon:
                return a
            elif abs(FB) < epsilon:
                return b
            # Verificando se o valor de f(x) é menor que epsilon e consequentemente a raiz
            elif abs(FX) < epsilon:
                return x
            # Escolha do intervalo
            elif FB > FA:
                b = x # Se f(b) for maior que f(a) então o intervalo é [a, x]
            else:
                a = x # Se f(a) for maior que f(b) então o intervalo é [x, b]
            iter_count_bisse += 1
            if iter_count_bisse > k: # Se o número de iterações for maior que k, então não foi possível encontrar a raiz
                return None
    else:
        # Para intervalos com f(a) e f(b) opostos
        while abs(b - a) > epsilon:
            x = (a + b) / 2
            if f(x) == 0:
                return x
            elif f(a) * f(x) < 0:
                b = x
            else:
                a = x
            iter_count_bisse += 1
            print(f"Iteração {iter_count_bisse}: x = {x}, f(x) = {f(x)}")
        return x



# Entrada de dados
a = float(input("Digite o valor de a: "))
b = float(input("Digite o valor de b: "))
epsilon = 0.01
k = 1000 # k para falsa posição

inicio = time.time() 
raiz = falsa_posicao(a, b, epsilon, k)
tempo_total_falsa = time.time() - inicio

inicio = time.time() 
kbisse = calcK(a, b, epsilon)
raizBisse = bisseccao(a, b, epsilon, kbisse)
tempo_total_bisse = time.time() - inicio

print("\n\n=========================\n")
print("Método da Falsa Posição")
print(f"Iterações: {iter_count}")
print(f"k máximo: {k}")
print(f"Tempo (segundos): {tempo_total_falsa}")


print("\n\n=========================\n")
print("Método da Bissecção")
print(f"Iterações: {iter_count_bisse}")
print(f"k máximo: {kbisse}")
print(f"Tempo (segundos): {tempo_total_bisse}")

print("\n\n=========================\n")
print("Raízes")

if raizBisse is not None:
    print(f"A raiz aproximada da bissecção: {raizBisse:.5f}.")
else:
    print(f"Não foi possível encontrar a raiz com os parâmetros fornecidos.")


if raiz is not None and raizBisse is not None:
    print(f"A raiz aproximada da falsa posição é: {raiz:.5f}.")
else:
    print(f"Não foi possível encontrar a raiz com os parâmetros fornecidos.")
```
