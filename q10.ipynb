```python
import math
import time

# Função f(x) da questão
L3 = 3
E = 70e9
I = 52.9e-6  
W0 = 15e3
y = 0.009  # Deflexão desejada em metros

def f(x):
    return (((W0*L)/(3*math.pi*4*E*I))) * (48*L3*math.cos((math.pi*x)/(2*L)) - 48*L3 + 3*math.pi3*L*x2 - math.pi3*x*3) - y

# Função pra derivada
def derivada(x, epsilon):
    """Calcula a derivada numérica da função f em x usando o valor de epsilon."""
    return (f(x + epsilon) - f(x)) / epsilon

# Definição Método de Newton
def metodo_newton(x0, epsilon, max_iter=100):
    """
    Implementa o método de Newton para encontrar a raiz da função f.

    Args:
    x0: Ponto inicial para o método de Newton.
    epsilon: Precisão desejada para a raiz.
    max_iter: Número máximo de iterações para evitar loops infinitos.

    Returns:
    A raiz encontrada da função f.
    """
    x = x0
    iter_count = 0

    while abs(f(x)) > epsilon and iter_count < max_iter:
        df = derivada(x, epsilon)
        if df == 0:
            raise ValueError("A derivada é zero. O método de Newton pode falhar.")

        x = x - f(x) / df
        iter_count += 1

        print(f"Iteração {iter_count}: x = {x}, f(x) = {f(x)}")
    print("--------------------------------------------------")
    print(f"Total de iterações: {iter_count}")

    return x

# Definição Método Secante
def metodo_secante(x0, x1, epsilon, max_iter=100):
    """
    Implementa o método da secante para encontrar a raiz da função f.

    Args:
    x0: Primeiro ponto inicial.
    x1: Segundo ponto inicial.
    epsilon: Precisão desejada para a raiz.
    max_iter: Número máximo de iterações para evitar loops infinitos.

    Returns:
    A raiz encontrada da função f.
    """
    iter_count = 0

    while abs(f(x1)) > epsilon and iter_count < max_iter:
        if f(x1) == f(x0):
            raise ValueError("Divisão por zero na iteração. Método falhou.")

        # Fórmula do método da secante
        x_new = x1 - f(x1) * (x1 - x0) / (f(x1) - f(x0))

        # Atualizar os pontos
        x0, x1 = x1, x_new
        iter_count += 1
        print(f"Iteração {iter_count}: x = {x1}, f(x) = {f(x1)}")

        if iter_count == max_iter:
            print("Número máximo de iterações atingido.")
            return None
    print("----------------------------------------------------------")
    print(f"Total de iterações: {iter_count}")
    return x1

# Precisão
epsilon = 0.001

# Entrada do usuário - Secante
x0 = 0
x1 = 1.5


# Solução Método da Secante
t_secante = time.time()
print("==========================================================")
print("Método da Secante: ")
raiz = metodo_secante(x0, x1, epsilon)
t_secante_final = time.time() - t_secante
print("X do método da secante:", raiz)
print("Tempo de execução (Secante): ", t_secante_final)

# Entrada do usuário - Newton
x0 = 1.5

# Solução Método de Newton
t_newton = time.time()
print("==========================================================")
print("Método de Newton: ")
raiz = metodo_newton(x0, epsilon)
t_newton_final = time.time() - t_newton
print("X do método de Newton:", raiz)
print("Tempo de execução (Newton): ", t_newton_final)
```
